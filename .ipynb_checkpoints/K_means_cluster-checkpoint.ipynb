## K-Means Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
```

```python
df = pd.read_csv(r'Mall_Customers.csv',index_col=0,header=0)
df
```

```python
df.describe(include = 'all')
```

```python
df['Gender'] = df.Gender.map({'Female':0,'Male':1})
```

```python
df
```

```python
df['Gender'].value_counts().plot(kind = 'pie',autopct='%.2f%%');
```

```python
X = df.values[:,[2,3]]
```

```python
X
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
 
wsse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,  random_state = 10)
    kmeans.fit(X)
    wsse.append(kmeans.inertia_)  # inertia_ gives the sum of squared distances of samples to their closest cluster center.
    
plt.plot(range(1, 11), wsse,c="blue")
plt.scatter(range(1, 11),wsse,c="red")
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSSE')
plt.show()
```

```python
print(wsse)
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 10)
Y_pred = kmeans.fit_predict(X)
```

```python
Y_pred
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
df
```

```python
df['Clusters'] = Y_pred
```

```python
df
```

```python
kmeans.cluster_centers_
```

```python
sns.lmplot(data=df, x="Annual Income (k$)",y="Spending Score (1-100)",
          fit_reg=False, #No regression line
          hue="Clusters",palette="Set1")
 
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
           s=300,c="yellow")
plt.show()
```

```python
df['Clusters'] = df.Clusters.map({0:'Standard',1:'Valuable',2:'Most Valuable',3:'Elite',4:'Carefull'})
```

```python
df
```

```python
df['Clusters'].value_counts().plot(kind = 'bar');
```

```python
new_df = df[df['Clusters'] == 'Elite']
new_df
```

```python
df
```

```python
X = df.values[:,0:-1]
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Clusters'] = le.fit_transform(df['Clusters'])
```

```python
Y = df.values[:,-1]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 10)
```

```python
x_train.shape
```

```python
#predicting using the Decision_Tree_Classifier
from sklearn.tree import DecisionTreeClassifier
 
model_DecisionTree = DecisionTreeClassifier(criterion="gini",random_state=10)
 
#fit the model on the data and predict the values
model_DecisionTree.fit(x_train,y_train)
 
Y_predict = model_DecisionTree.predict(x_test)
#print(Y_pred)
#print(list(zip(Y_test,Y_pred)))
```

```python

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
 
cfm=confusion_matrix(y_test,Y_predict)

sns.heatmap(cfm, annot = True, fmt = 'g', cbar = False ,cmap = 'BuPu')
plt.xlabel('Predicted value')
plt.ylabel('Actual value')
plt.show()

print(cfm)
 
print("Classification report: ")
 
print(classification_report(y_test,Y_predict))
 
acc=accuracy_score(y_test,Y_predict)
print("Accuracy of the model: ",acc)
```

```python
model_DecisionTree.score(x_train,y_train)
```

```python
model_DecisionTree.score(x_test,y_test)
```

```python
import gradio as gr
import numpy as np
import joblib
import os

# Load model safely
MODEL_PATH = "./model.pkl"
ENCODER_PATH = "./label_encoders.pkl"

model = joblib.load(MODEL_PATH)
cluster_map = joblib.load(ENCODER_PATH)

def predict_cluster(gender, age, income, spending):
    gender = 0 if gender == "Female" else 1
    X = np.array([[gender, age, income, spending]])

    cluster_id = int(model.predict(X)[0])

    # inverse transform
    cluster_name = le.inverse_transform([cluster_id])[0]

    return f"Customer Segment: {cluster_name}"

with gr.Blocks() as demo:
    gr.Markdown("## 🧠 Customer Segmentation (K-Means)")

    gender = gr.Radio(["Female", "Male"], label="Gender")
    age = gr.Number(label="Age")
    income = gr.Number(label="Annual Income (k$)")
    spending = gr.Number(label="Spending Score (1-100)")

    output = gr.Textbox(label="Result")
    btn = gr.Button("Predict")

    btn.click(
        fn=predict_cluster,
        inputs=[gender, age, income, spending],
        outputs=output
    )

# IMPORTANT: queue बंद
demo.launch(share=False, debug=True)
```
